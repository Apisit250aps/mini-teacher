```python
import requests
from faker import Faker

fake = Faker('th_TH')
```

```python

for i in range(1, 20):
    data = {
        "code": f"62106332{i:02}",
        "teacher": "019c66fc-2b07-739f-869c-cc689fe27b13",
        "firstName": fake.first_name(),
        "lastName": fake.last_name(),
        "prefix": fake.prefix(),
    }
    print(data)
    response = requests.post(
        'http://localhost:3000/api/student', json=data)
    print(response.status_code, response.json())
```
