# Mini Teacher Seed Data

## Set up

```python
import requests
import math
from faker import Faker


baseUrl = 'http://localhost:3000'
fake = Faker('th_TH')
```

## Students seed

```python
n = 10
for i in range(1, n + 1):
    is_male = i < math.ceil(n / 2)
    data = {
        "code": f"65106332{i:02}",
        "teacherId": "40b4a0cb-bd9f-4d04-9ddb-393d2c1e2170",
        "firstName": fake.first_name_male() if is_male else fake.first_name_female(),
        "lastName": fake.last_name(),
        "prefix": "นาย" if is_male else "นาวสาว",
    }
    student = requests.post(
        f'{baseUrl}/api/student', json=data)

    classMember = requests.post(
        f'{baseUrl}/api/class-member', json={
            "classId": "51553b62-350b-4ecf-ab5a-7bc8e2ed0013",
            "studentId": student.json()['data']['id']
        })
```
